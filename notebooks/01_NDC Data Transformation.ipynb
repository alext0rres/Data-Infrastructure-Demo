```python
from pyspark.sql import functions as F

# load in raw ndc data
ndc_raw = (
    spark.read
         .option("multiLine", True)
         .json("/Volumes/workspace/default/abbvie_demo_files/drug-ndc-0001-of-0001.json")
)

# inspect dataset schema
ndc_raw.printSchema()
```

```python
# access results array and flatten into dataframe
ndc = ndc_raw.select(F.explode("results").alias("r"))
ndc_df = ndc.select("r.*")
display(ndc_df.limit(10))
```

```python
# display ndc_df columns
display(ndc_df.columns)
```

```python
# create drug_products table from ndc_df
drug_products = (
    ndc_df.select(
        F.col("product_ndc"),
        F.col("brand_name"),
        F.col("labeler_name"),
        F.col("generic_name"),
        F.col("dosage_form"),
        F.col("product_type")
    )
    .dropDuplicates(["product_ndc"])
)
```

```python
# create product_ingredients table from ndc_df
product_ingredients = (
    ndc_df.select(
        F.col("product_ndc"),
        F.explode("active_ingredients").alias("ingredient_obj")
    )
    .select(
        "product_ndc",
        F.upper(F.trim(F.col("ingredient_obj.name"))).alias("ingredient_name"),
        F.col("ingredient_obj.strength").alias("ingredient_strength"),
    )
    .dropDuplicates()
)
```

```python
# save drug_products and product_ingredients as tables
drug_products.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("drug_products")
product_ingredients.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("product_ingredients")
```

```python
# check row count for each table
drug_count = spark.sql("SELECT COUNT(*) AS drug_count FROM drug_products").collect()[0]["drug_count"]
print(f"drug_products row count: {int(drug_count)}")

ingredient_count = spark.sql("SELECT COUNT(*) AS ingredient_count FROM product_ingredients").collect()[0]["ingredient_count"]
print(f"product_ingredients row count: {int(ingredient_count)}")
```

```python
# inspect data
display(spark.sql("SELECT * FROM drug_products LIMIT 10"))
display(spark.sql("SELECT * FROM product_ingredients LIMIT 10"))
```

```python
# create ingredient dimension
ingredient_dim = (
    spark.sql("SELECT DISTINCT ingredient_name FROM product_ingredients")
)

# save ingredient_dim as table and display
ingredient_dim.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("ingredients")
display(spark.sql("SELECT * FROM ingredients LIMIT 10"))
```
