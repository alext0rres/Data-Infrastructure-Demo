```python
from pyspark.sql import functions as F

# load in raw faers data
faers_raw = (
    spark.read
    .option("multiline", True)
    .json("/Volumes/workspace/default/abbvie_demo_files/drug-event-0001-of-0033.json")
)

# inspect dataset schema
faers_raw.printSchema()
```

```python
# access results array and flatten into dataframe
faers = faers_raw.select(F.explode("results").alias("r"))
faers_df = faers.select("r.*")
display(faers_df.limit(10))
```

```python
# create "reports" table with identifiable adverse event report information (choose columns based on schema)
reports = (
    faers_df.select(
        F.col("safetyreportid").alias("report_id"),
        F.to_date(F.col("receivedate"), "yyyyMMdd").alias("report_date"),
        F.col("serious"),
        F.col("occurcountry")
    )
    .dropna(subset=["report_id"])
    .dropDuplicates(["report_id"])
)

display(reports.limit(10))
```

```python
# create "report_events" table with specific adverse event information (choose columns based on schema)
report_events = (
    faers_df.select(
        F.col("safetyreportid").alias("report_id"),
        F.explode_outer("patient.reaction").alias("reaction_obj")
    )
    .select(
        "report_id",
        F.upper(F.trim(F.col("reaction_obj.reactionmeddrapt"))).alias("event_name"),
        F.col("reaction_obj.reactionoutcome").alias("event_outcome")
    )
    .dropna(subset=["report_id", "event_name"])
    .dropDuplicates()
)

display(report_events.limit(10))
```

```python
# create "report_drug" normalization table with openFDA drug information stored as arrays - will serve as basis for later post-staging bridge table creation done by exploding the relevant nested openFDA arrays into their respective tables. Each bridge table will represent a many-to-many relationship and act as an intermediate relational layer before eventual conversion to graph.
report_drug = (
    faers_df.select(
        F.col("safetyreportid").alias("report_id"),
        F.explode_outer("patient.drug").alias("drug_obj")
    )
    .select(
        "report_id",
        F.upper(F.trim(F.col("drug_obj.medicinalproduct"))).alias("drug_name"),
        F.col("drug_obj.openfda.product_ndc").alias("product_ndc_arr"),
        F.col("drug_obj.openfda.rxcui").alias("rxcui_arr"),
        F.col("drug_obj.openfda.generic_name").alias("generic_name_arr"),
        F.col("drug_obj.openfda.brand_name").alias("brand_name_arr"),
        F.col("drug_obj.openfda.manufacturer_name").alias("manufacturer_name_arr"),
        F.col("drug_obj.openfda.substance_name").alias("substance_name_arr"),
        F.col("drug_obj.drugcharacterization").alias("drug_characterization")
    )
)

display(report_drug.limit(10))
```

```python
# save reports, report_events, and report_drug as tables
reports.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("reports")
report_events.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("report_events")
report_drug.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("report_drug")
```

```python
# identify unique coded categorical values for report_events.event_outcome, reports.serious, and report_drug.drug_characterization
report_events.select("event_outcome").distinct().show()
reports.select("serious").distinct().show()
report_drug.select("drug_characterization").distinct().show()
```

```python
# check null and total row counts for event_outcome, serious, and drugcharacterization
event_outcome_nulls = report_events.filter(F.col("event_outcome").isNull()).count()
serious_nulls = reports.filter(F.col("serious").isNull()).count()
drugcharacterization_nulls = report_drug.filter(F.col("drug_characterization").isNull()).count()

total_report_events = report_events.count()
total_reports = reports.count()
total_dc = report_drug.count()

print(f'Total rows in report_events: {total_report_events}')
print(f'Total nulls in event_outcome: {event_outcome_nulls}')
print()
print(f'Total rows in reports: {total_reports}')
print(f'Total nulls in serious: {serious_nulls}')
print()
print(f'Total rows in report_drug: {total_dc}')
print(f'Total nulls in drug_characterization: {drugcharacterization_nulls}')
```

```python
# cast "serious" and "event_outcome" values to int before mapping and verify no additional null values
reports = reports.withColumn(
    "serious",
    F.col("serious").cast("int")
)
report_events = report_events.withColumn(
    "event_outcome",
    F.col("event_outcome").cast("int")
)
report_drug = report_drug.withColumn(
    "drug_characterization",
    F.col("drug_characterization").cast("int")
)

print(f'Total nulls in event_outcome: {event_outcome_nulls}')
print(f'Total nulls in serious: {serious_nulls}')
print(f'Total nulls in drug_characterization: {drugcharacterization_nulls}')
```

```python
# create mapping layer to translate reports.serious, report_events.event_outcome, and report_drug.drug_characterization values into string labels based on FAERS documentation

reports = reports.withColumn(
    "serious_label",
    F.when(F.col("serious") == 1, "SERIOUS")
    .when(F.col("serious") == 2, "NOT_SERIOUS")
    .otherwise("UNKNOWN")
)

report_events = report_events.withColumn(
    "event_outcome_label",
    F.when(F.col("event_outcome") == 1, "RECOVERED")
    .when(F.col("event_outcome") == 2, "RECOVERING")
    .when(F.col("event_outcome") == 3, "NOT_RECOVERED")
    .when(F.col("event_outcome") == 4, "RECOVERED_WITH_SEQUELAE")
    .when(F.col("event_outcome") == 5, "FATAL")
    .when(F.col("event_outcome") == 6, "UNKNOWN")
    .otherwise("UNKNOWN")
)

report_drug = report_drug.withColumn(
    "drug_characterization_label",
    F.when(F.col("drug_characterization") == 1, "PRIMARY_SUSPECT_DRUG")
    .when(F.col("drug_characterization") == 2, "SECONDARY_SUSPECT_DRUG")
    .when(F.col("drug_characterization") == 3, "CONCOMITANT_DRUG")
    .when(F.col("drug_characterization") == 4, "INTERACTING_DRUG")
)

# confirm correct mappings
display(reports.select("serious", "serious_label").limit(5))
display(report_events.select("event_outcome", "event_outcome_label").limit(5))
display(report_drug.select("drug_characterization", "drug_characterization_label").limit(5))
```

```python
# resave tables to reflect mapping changes
reports.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("reports")
report_events.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("report_events")
report_drug.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("report_drug")
```
